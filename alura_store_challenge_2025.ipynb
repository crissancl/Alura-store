```python
import pandas as pd
```

```python
df1 = pd.read_csv('tienda_1 .csv')  # con el espacio antes del punto
df2 = pd.read_csv('tienda_2.csv')
df3 = pd.read_csv('tienda_3.csv')
df4 = pd.read_csv('tienda_4.csv')
```

```python
# Facturación total (precio * cantidad)
fact1 = (df1['Precio'] * df1['Cantidad de cuotas']).sum()
fact2 = (df2['Precio'] * df2['Cantidad de cuotas']).sum()
fact3 = (df3['Precio'] * df3['Cantidad de cuotas']).sum()
fact4 = (df4['Precio'] * df4['Cantidad de cuotas']).sum()

# Mostrar resultados
print('Facturación total tienda 1:', fact1)
print('Facturación total tienda 2:', fact2)
print('Facturación total tienda 3:', fact3)
print('Facturación total tienda 4:', fact4)
```

```python
# Categorías más vendidas por tienda
print("Categorías más populares por tienda:\n")

print("Tienda 1:")
print(df1['Categoría del Producto'].value_counts().head(3), "\n")

print("Tienda 2:")
print(df2['Categoría del Producto'].value_counts().head(3), "\n")

print("Tienda 3:")
print(df3['Categoría del Producto'].value_counts().head(3), "\n")

print("Tienda 4:")
print(df4['Categoría del Producto'].value_counts().head(3), "\n")
```

```python
import matplotlib.pyplot as plt

# Nombres de tiendas y facturaciones
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
facturaciones = [fact1, fact2, fact3, fact4]

# Crear gráfico
plt.figure(figsize=(8, 5))
plt.bar(tiendas, facturaciones)
plt.title('Facturación Total por Tienda')
plt.ylabel('CLP')
plt.xlabel('Tienda')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
